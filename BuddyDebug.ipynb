# Code Runner

```python
def run_user_code(user_code):
    try:
        local_vars = {}
        from io import StringIO
        import sys


        old_stdout = sys.stdout
        sys.stdout = mystdout = StringIO()


        exec(user_code, {}, local_vars)


        sys.stdout = old_stdout
        output = mystdout.getvalue()

        if output.strip() == "":
            return "‚úåÔ∏èCode ran successfully (no output)."
        else:
            return "üòåOutput:\n" + output.strip()

    except Exception as e:
        # Restore stdout
        sys.stdout = old_stdout
        return f"üò° Error: {type(e).__name__}: {str(e)}"
```

# Auto Fixer

```python
def auto_fix_code(user_code, error_message):
    """
    Takes user code and error message, returns a suggested fixed code snippet.
    """
    prompt = (
        "You are a Python coding assistant. "
        "The user wrote some code and got an error. "
        "Suggest a corrected version of the code, fix only what is necessary, "
        "and explain briefly what you fixed.\n\n"
        f"Code:\n{user_code}\n\n"
        f"Error:\n{error_message}\n\n"
        "Return only the corrected code and a short explanation."
    )

    # Call Gemini or OpenAI here
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text
```

```python
def buddy_debug(user_code):
    # Part 1: Run code
    output = run_user_code(user_code)

    # Part 2: AI explanation
    _, explanation = explain_code(user_code)

    # Part 3: Auto-Fixer
    autofixed = auto_fix_code(user_code, output)

    return output, explanation, autofixed
```

# BuddyDebug v1.1

```python
def buddy_debug(user_code):
    # Part 1: Run code
    output = run_user_code(user_code)

    # Part 2: AI explanation
    _, explanation = explain_code(user_code)

    # Check if there's an error in the output
    if "Error" in output or "Traceback" in output or "Exception" in output:
        # Part 3: Auto-Fixer
        autofixed = auto_fix_code(user_code, output)

        # Part 4: Practice Maker
        genprac = generate_practice(user_code, output)
    else:
        # No error ‚Äî skip fixing and practice generation
        autofixed = "‚úÖ No issues detected! Your code ran successfully."
        genprac = "üéâ Great job! No extra exercises needed."

    return output, explanation, autofixed, genprac
```

```python
code = """
print('it is working!')
"""
output, explanation, autofixed, genprac = buddy_debug(code)

print("=== Runner Output ===")
print(output)

print("\n=== AI Explanation ===")
print(explanation)

print("\n=== Auto Fixer ===")
print(autofixed)

print("\n=== New Problems ===")
print(genprac)
```

# BuddyDebug v1.2

```python
import sys, time

def typewriter(text, delay=0.06):
    """
    Prints text like a typing animation.
    """
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(delay)
    print()  # new line after finishing
```

```python
code = "print(hello)"
output, explanation, autofixed, genprac = buddy_debug(code)

print("=== Runner Output ===")
typewriter(output)

print("\n=== AI Explanation ===")
typewriter(explanation)

print("\n=== Auto Fixer ===")
typewriter(autofixed)

print("\n=== New Problems ===")
typewriter(genprac)
```
